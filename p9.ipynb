```python
df_dispo_ali=dispo_ali.pivot_table(index=["Zone","Produit"],columns="Élément",values="Valeur",aggfunc="sum")
df_dispo_ali.head()
```

```python
df_dispo_ali_col=df_dispo_ali[["Disponibilité intérieure",
                              "Exportations - Quantité","Importations - Quantité","Nourriture",
                              "Pertes","Production"]]
df_dispo_ali_col.head()
```

```python
df_dispo_ali_col.reset_index(inplace=True)
df_dispo_ali_col.head()
```

```python
df_dispo_ali_col.rename_axis(columns=None,inplace=True)
df_dispo_ali_col.head()
```

```python
df_dispo_ali_col.Produit.unique()
```

```python
volaille=["Viande de Volailles"]
autres_viandes=["Viande d'Ovins/Caprins", 'Viande de Bovins', 'Viande de Suides']
cereales=['Céréales, Autres','Millet et produits']
```

```python
def type_produit (produit):
    if produit in volaille : 
        return "volaille"
    if produit in autres_viandes : 
        return "autres viandes"
    if produit in cereales : 
        return "cereales"
    else :
        return "autres"
```

```python
df_dispo_ali_col["Type_produit"]=df_dispo_ali_col.Produit.apply(type_produit)
df_dispo_ali_col.head(50)
```

```python
df_dispo_semifinal=df_dispo_ali_col[df_dispo_ali_col["Type_produit"]!="autres"].groupby(["Zone","Type_produit"]).sum()
df_dispo_semifinal.head()
```

```python
df_seminifal=df_dispo_semifinal.pivot_table(index="Zone",columns=["Type_produit"],
                               values=["Disponibilité intérieure","Exportations - Quantité","Importations - Quantité",
                                       "Nourriture","Pertes","Production"],aggfunc="sum")
df_seminifal.head()
```

```python
df_seminifal.reset_index(inplace=True)
df_seminifal.head()
```

```python
df_seminifal.rename_axis(columns=None,inplace=True)
df_seminifal.head()
```

```python
df_final=pd.merge(df_seminifal,population[population["Année"]==2017][["Zone","Valeur"]],on="Zone",how="left")
df_final.head()
```
